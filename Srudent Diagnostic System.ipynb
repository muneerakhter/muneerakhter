```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
class StudentDiagnosticSystem:
    def __init__(self):
        # Data transcribed from Table 1 in the assignment image
        self.data = {
            'ATQ': [500, 550, 450, 400, 600, 650, 700, 550, 650, 700, 550, 650, 550],
            'ILEP': [30, 32, 28, 25, 32, 38, 39, 38, 38, 39, 38, 35, 31],
            'GPA': [2.8, 3.0, 2.9, 2.8, 3.3, 3.3, 3.5, 3.7, 3.3, 3.5, 3.7, 3.4, 2.9]
        }
        self.df = pd.DataFrame(self.data)
        self.model = None
        self.results = None
    def display_data_summary(self):
        print("--- Data Snapshot ---")
        print(self.df.head())
        print("\n--- Correlation Matrix ---")
        print(self.df.corr())
    def train_model(self):
        """
        Fits a Multiple Linear Regression model using OLS (Ordinary Least Squares).
        """
        X = self.df[['ATQ', 'ILEP']]
        y = self.df['GPA']
        # Add a constant (intercept) to the predictors
        X_with_const = sm.add_constant(X)
        # Fit the model
        self.model = sm.OLS(y, X_with_const).fit()
        self.results = self.model
        print("\n--- Regression Model Results ---")
        print(self.model.summary())
    def analyze_partial_association(self):
        """
        Addresses Part C: Checking association between GPA and ILEP
        while controlling for ATQ (Partial Correlation).
        """
        # Calculate residuals of GPA ~ ATQ
        model_gpa_atq = sm.OLS(self.df['GPA'], sm.add_constant(self.df[['ATQ']])).fit()
        residuals_gpa = model_gpa_atq.resid
        # Calculate residuals of ILEP ~ ATQ
        model_ilep_atq = sm.OLS(self.df['ILEP'], sm.add_constant(self.df[['ATQ']])).fit()
        residuals_ilep = model_ilep_atq.resid
        # Correlation between residuals represents the partial correlation
        partial_corr = np.corrcoef(residuals_gpa, residuals_ilep)[0, 1]
        print("\n--- Part C Analysis: Partial Correlation ---")
        print(f"Partial Correlation (GPA vs ILEP | ATQ): {partial_corr:.4f}")
        if abs(partial_corr) < 0.1:
            print("Interpretation: Once ATQ is controlled, ILEP has negligible association with GPA.")
        else:
            print("Interpretation: ILEP maintains an association with GPA even after controlling for ATQ.")
# --- Execution ---
if __name__ == "__main__":
    system = StudentDiagnosticSystem()
    system.display_data_summary()
    system.train_model()
    system.analyze_partial_association()
```
